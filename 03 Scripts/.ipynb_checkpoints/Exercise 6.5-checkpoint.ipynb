### 2. The elbow technique

```python
num_cl = range(1, 10) # Defines the range of potential clusters in the data.
kmeans = [KMeans(n_clusters=i) for i in num_cl] # Defines k-means clusters in the range assigned above.
```

```python
# Drop non-numeric columns
numeric_df = df.select_dtypes(include='number')
```

```python
score = [kmeans[i].fit(numeric_df).score(numeric_df) for i in range(len(kmeans))] # Creates a score that represents 
# a rate of variation for the given cluster option.

score
```

```python
# Plot the elbow curve using PyLab.

pl.plot(num_cl,score)
pl.xlabel('Number of Clusters')
pl.ylabel('Score')
pl.title('Elbow Curve')
pl.show()
```

#####  There is a large decrease in the score when moving from two to three clusters, indicating that adding one more cluster significantly improves the clustering performance. However, the decrease in the score from three to four clusters is less pronounced, and the curve starts to flatten out after that. Based on this observation, it seems that the optimal number of clusters for your data is four.

### 3. k-means clustering

```python
# Create the k-means object.

kmeans = KMeans(n_clusters = 4) 
```

```python
# Fit the k-means object to the data.

kmeans.fit(numeric_df)
```

```python
df['clusters'] = kmeans.fit_predict(numeric_df)
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df['clusters'].value_counts()
```

```python
# Plot the clusters for the "Adult Mortality" and "Alcohol" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Adult Mortality'], y=df['Alcohol'], hue=kmeans.labels_, s=20) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Adult Mortality') # Label x-axis.
plt.ylabel('Alcohol') # Label y-axis.
plt.show()
```

```python
# Plot the clusters for the "Life Expectancy" and "GDP" variables.

plt.figure(figsize=(12,8))
ax = sns.scatterplot(x=df['Life Expectancy'], y=df['GDP'], hue=kmeans.labels_, s=20) 
# Here, you're subsetting `X` for the x and y arguments to avoid using their labels. 
# `hue` takes the value of the attribute `kmeans.labels_`, which is the result of running the k-means algorithm.
# `s` represents the size of the points you want to see in the plot.

ax.grid(False) # This removes the grid from the background.
plt.xlabel('Life Expectancy') # Label x-axis.
plt.ylabel('GDP') # Label y-axis.
plt.show()
```

```python
df.loc[df['clusters'] == 3, 'cluster'] = 'black'
df.loc[df['clusters'] == 2, 'cluster'] = 'dark purple'
df.loc[df['clusters'] == 1, 'cluster'] = 'purple'
df.loc[df['clusters'] == 0, 'cluster'] = 'pink'
```

```python
df.groupby('cluster').agg({'Adult Mortality':['mean', 'median'], 
                         'Alcohol':['mean', 'median'], 
                         'Life Expectancy':['mean', 'median'],
                          'GDP':['mean', 'median'],
                          'Income Composition':['mean',  'median'],
                          'BMI':['mean', 'median'],
                          'Schooling':['mean', 'median']})
```
